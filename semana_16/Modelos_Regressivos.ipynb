## Importando os pacotes

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### leitura do dataset carros
carros = pd.read_csv("carprices.csv")
```

```python
### leitura das 5 primeiras linhas
carros.head()
```

```python
### analise de descritivas
carros.describe()
```

## AnÃ¡lise de correlaÃ§Ã£o

```python
### seleciona variaveis numericas
num_carros = carros.select_dtypes(include=['float64', 'int64'])
## calcula a matriz de correlaÃ§Ã£o
corr = num_carros.corr()
```

```python
# Plot
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlacao')
plt.show()
```

## Analise dos grÃ¡ficos de dispersÃ£o

```python
### iterando em todas as colunas numÃ©ricas
for column in num_carros.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=num_carros, x=column, y="price")
    plt.title(f'GrÃ¡fico de dispersÃ£o {column} vs PreÃ§o')
    plt.xlabel(column)
    plt.ylabel("preÃ§o")
    plt.show()
```

Notamos que algumas variÃ¡veis parecem ter uma relaÃ§Ã£o bem linear com o preÃ§o.
Vamos agora aplicar os conceitos de regressÃ£o linear simples

## Regressao linear simples: PreÃ§o x Enginesize

Vamos considerar uma regressao linear simples de preÃ§o e engine size. variÃ¡veis com correlaÃ§ao de 0.86 pela matriz e grÃ¡fico de dispersÃ£o promissor

```python
### importa o pacote
import statsmodels.api as sm

### define a variÃ¡vel explicativa e adiciona uma constante
X = sm.add_constant(num_carros['enginesize'])


### define a variÃ¡vel de interesse

y = num_carros['price']
```

```python
# Fit do modelo: aqui o python ira calcular os valores de intercepto e coeficiente angular estimados
model = sm.OLS(y, X).fit()
```

```python
## gera a tabela de regressÃ£o
print(model.summary())
```

O modelo estimado foi:

PreÃ§o do carro = -8005.445 + 167.6984*(tamanho do carro)

```python
## gerando o grÃ¡fico da reta estimada

plt.figure(figsize=(8, 6))
sns.scatterplot(data=num_carros, x='enginesize', y='price', color='blue', alpha=0.5)

# a funcao regplot gera a regressao e a plot da reta
sns.regplot(data=num_carros, x='enginesize', y='price', scatter=False, color='red')

# definindo titulos
plt.title('RegressÃ£o do PreÃ§o do VeÃ­culo vs Tamanho')
plt.xlabel('Tamanho')
plt.ylabel('PreÃ§o')

plt.show()
```

Vamos agora analisar os resÃ­duos do modelo:

```python
# salva o dado do resÃ­duo
residuals = model.resid
# calcula ovalor predito
predicted_values = model.fittedvalues

plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, color='blue', alpha=0.5)

# adiciona linha
plt.axhline(y=0, color='red', linestyle='--')

# titulos
plt.title('Residuo vs Valor Predito')
plt.xlabel('Valor predito')
plt.ylabel('Residuo')

plt.show()
```

## RegressÃ£o Multipla

Vamos agora construir um modelo regressÃ£o multipla.

Para isso vamos considerar as variÃ¡veis numÃ©ricas de correlaÃ§Ã£o forte

```python
### seleciona variÃ¡veis com correlacao elevada
high_correlation_variables = corr[(corr['price'] > 0.7) | (corr['price'] < -0.7)].index.tolist()
high_correlation_variables.remove('price')

# variÃ¡veis selecionadas
selected_variables = ['price'] + high_correlation_variables
selected_df = num_carros[selected_variables]
```

```python
# Fit do modelo
X = sm.add_constant(selected_df.drop(columns=['price']))
y = selected_df['price']
model = sm.OLS(y, X).fit()

# Print regression summary
print(model.summary())
```

Vemos que a variÃ¡vel curbweight nÃ£o tem p-valor menor que 0.05.
Assim vamos retira-la do modelo. esse processo pode ser feito iterativamente, atÃ© mantermos variÃ¡veis significativas

```python
high_correlation_variables.remove('curbweight')

# variÃ¡veis re-selecionadas
selected_variables = ['price'] + high_correlation_variables
selected_df = num_carros[selected_variables]
```

```python
# Fit do modelo atualizado
X = sm.add_constant(selected_df.drop(columns=['price']))
y = selected_df['price']
model = sm.OLS(y, X).fit()

# Print regression summary
print(model.summary())
```

O modelo estimado foi:

PreÃ§o do carro = -6.021e-4 + 848.69*(largura do carro ) + 94.94*(tamanho do carro ) +  52.8026*(potencia do carro)

AnÃ¡lise dos resÃ­duos

```python
# salva o dado do resÃ­duo
residuals = model.resid
# calcula ovalor predito
predicted_values = model.fittedvalues

plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, color='blue', alpha=0.5)

# adiciona linha
plt.axhline(y=0, color='red', linestyle='--')

# titulos
plt.title('Residuo vs Valor Predito')
plt.xlabel('Valor predito')
plt.ylabel('Residuo')

plt.show()
```
