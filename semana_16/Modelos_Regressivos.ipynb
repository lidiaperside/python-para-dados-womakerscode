## Importando os pacotes

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### leitura do dataset carros
carros = pd.read_csv("carprices.csv")
```

```python
### leitura das 5 primeiras linhas
carros.head()
```

```python
### analise de descritivas
carros.describe()
```

## Análise de correlação

```python
### seleciona variaveis numericas
num_carros = carros.select_dtypes(include=['float64', 'int64'])
## calcula a matriz de correlação
corr = num_carros.corr()
```

```python
# Plot
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlacao')
plt.show()
```

## Analise dos gráficos de dispersão

```python
### iterando em todas as colunas numéricas
for column in num_carros.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=num_carros, x=column, y="price")
    plt.title(f'Gráfico de dispersão {column} vs Preço')
    plt.xlabel(column)
    plt.ylabel("preço")
    plt.show()
```

Notamos que algumas variáveis parecem ter uma relação bem linear com o preço.
Vamos agora aplicar os conceitos de regressão linear simples

## Regressao linear simples: Preço x Enginesize

Vamos considerar uma regressao linear simples de preço e engine size. variáveis com correlaçao de 0.86 pela matriz e gráfico de dispersão promissor

```python
### importa o pacote
import statsmodels.api as sm

### define a variável explicativa e adiciona uma constante
X = sm.add_constant(num_carros['enginesize'])


### define a variável de interesse

y = num_carros['price']
```

```python
# Fit do modelo: aqui o python ira calcular os valores de intercepto e coeficiente angular estimados
model = sm.OLS(y, X).fit()
```

```python
## gera a tabela de regressão
print(model.summary())
```

O modelo estimado foi:

Preço do carro = -8005.445 + 167.6984*(tamanho do carro)

```python
## gerando o gráfico da reta estimada

plt.figure(figsize=(8, 6))
sns.scatterplot(data=num_carros, x='enginesize', y='price', color='blue', alpha=0.5)

# a funcao regplot gera a regressao e a plot da reta
sns.regplot(data=num_carros, x='enginesize', y='price', scatter=False, color='red')

# definindo titulos
plt.title('Regressão do Preço do Veículo vs Tamanho')
plt.xlabel('Tamanho')
plt.ylabel('Preço')

plt.show()
```

Vamos agora analisar os resíduos do modelo:

```python
# salva o dado do resíduo
residuals = model.resid
# calcula ovalor predito
predicted_values = model.fittedvalues

plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, color='blue', alpha=0.5)

# adiciona linha
plt.axhline(y=0, color='red', linestyle='--')

# titulos
plt.title('Residuo vs Valor Predito')
plt.xlabel('Valor predito')
plt.ylabel('Residuo')

plt.show()
```

## Regressão Multipla

Vamos agora construir um modelo regressão multipla.

Para isso vamos considerar as variáveis numéricas de correlação forte

```python
### seleciona variáveis com correlacao elevada
high_correlation_variables = corr[(corr['price'] > 0.7) | (corr['price'] < -0.7)].index.tolist()
high_correlation_variables.remove('price')

# variáveis selecionadas
selected_variables = ['price'] + high_correlation_variables
selected_df = num_carros[selected_variables]
```

```python
# Fit do modelo
X = sm.add_constant(selected_df.drop(columns=['price']))
y = selected_df['price']
model = sm.OLS(y, X).fit()

# Print regression summary
print(model.summary())
```

Vemos que a variável curbweight não tem p-valor menor que 0.05.
Assim vamos retira-la do modelo. esse processo pode ser feito iterativamente, até mantermos variáveis significativas

```python
high_correlation_variables.remove('curbweight')

# variáveis re-selecionadas
selected_variables = ['price'] + high_correlation_variables
selected_df = num_carros[selected_variables]
```

```python
# Fit do modelo atualizado
X = sm.add_constant(selected_df.drop(columns=['price']))
y = selected_df['price']
model = sm.OLS(y, X).fit()

# Print regression summary
print(model.summary())
```

O modelo estimado foi:

Preço do carro = -6.021e-4 + 848.69*(largura do carro ) + 94.94*(tamanho do carro ) +  52.8026*(potencia do carro)

Análise dos resíduos

```python
# salva o dado do resíduo
residuals = model.resid
# calcula ovalor predito
predicted_values = model.fittedvalues

plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, color='blue', alpha=0.5)

# adiciona linha
plt.axhline(y=0, color='red', linestyle='--')

# titulos
plt.title('Residuo vs Valor Predito')
plt.xlabel('Valor predito')
plt.ylabel('Residuo')

plt.show()
```
