```python
#1. Faça o uploado do CSV carros.csv
df_carros = pd.read_csv('carros.csv')
```

```python
#4. Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros
df_carros.sort_values('Quant Carros', ascending = False)
```

```python
#5. Descubra qual a proporção de carro para a população de cada estado e insira uma nova coluna
proporcao = df_carros['Quant Carros'] / df_carros['População do Estado']
proporcao
```

```python
df_carros['Carros por habitante'] = proporcao
df_carros.head()
```

```python
#6. Faça um dataframe com Estado e proporção de carro por habitante
df_estado_hab = df_carros[['Estado', 'Carros por habitante']]
df_estado_hab.head()
```
