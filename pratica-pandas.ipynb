```python
import pandas as pd
```

```python
#1. Faça o uploado do CSV carros.csv
df_carros = pd.read_csv('carros.csv')
```

```python
df_carros.index
```

```python
#4. Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros
df_carros.sort_values('Quant Carros', ascending = False)
```

```python
#5. Descubra qual a proporção de carro para a população de cada estado e insira uma nova coluna
proporcao = df_carros['Quant Carros'] / df_carros['População do Estado']
proporcao
```

```python
df_carros['Carros por habitante'] = proporcao
df_carros.head()
```

```python
#6. Faça um dataframe com Estado e proporção de carro por habitante
df_estado_hab = df_carros[['Estado', 'Carros por habitante']]
df_estado_hab.head()
```

utilizando o arquivo: dados_concurso.csv

7. Quantas pessoas são do estado do Piauí?
8. Qual a distribuição de escolaridade por região?
9. Qual é a porcentagem de pessoas com deficiência?

```python
df_concurso = pd.read_csv('dados_concurso.csv')
df_concurso.head()
```

```python
#7. Quantas pessoas são do estado do Piauí?
df_concurso[df_concurso['Estado'] == 'PI']['Número de Inscrição'].count()
```

```python
#8. Qual a distribuição de escolaridade por região?
df_concurso[['Número de Inscrição', 'Estado']].groupby('Estado').count()
```

```python
df_concurso[['Estado', 'Escolaridade', 'Número de Inscrição']].groupby(['Estado', 'Escolaridade']).count()
```

```python
#9. Qual é a porcentagem de pessoas com deficiência?
df_concurso['Deficiência'].value_counts(normalize=True) *100
```
