2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]


```python
print('Média das horas de sono por profissão: ')
df_saude.groupby('Profissão')['Duração do sono'].mean()
```

```python
print('Mediana das horas de sono por profissão: ')
df_saude.groupby('Profissão')['Duração do sono'].median()
```

```python
print('Moda das horas de sono por profissão: ')
df_saude.groupby('Profissão')['Duração do sono'].agg(pd.Series.mode)
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)


```python
df_saude.head()
```

```python
df_saude[['Pressão Sistólica', 'Pressão Diastólica']] = df_saude['Pressão Sanguínea'].str.split('/', expand=True)
```

```python
df_saude.head()
```

```python
df_saude['Pressão Sistólica'] = pd.to_numeric(df_saude['Pressão Sistólica'], errors = 'coerce')
df_saude['Pressão Diastólica'] = pd.to_numeric(df_saude['Pressão Diastólica'], errors = 'coerce')

df_saude.head()
```

```python
media_pressao_homens_sistolica = df_saude[df_saude['Gênero'] == 'Homem']['Pressão Sistólica'].mean()
media_pressao_homens_diastolica = df_saude[df_saude['Gênero'] == 'Homem']['Pressão Diastólica'].mean()

media_pressao_mulheres_sistolica = df_saude[df_saude['Gênero'] == 'Mulher']['Pressão Sistólica'].mean()
media_pressao_mulheres_diastolica = df_saude[df_saude['Gênero'] == 'Mulher']['Pressão Diastólica'].mean()
```

```python
print ('A média da pressão dos homens é de: ', (media_pressao_homens_sistolica + media_pressao_homens_diastolica) / 2)
print ('A média da pressão das mulheres é de: ', (media_pressao_mulheres_sistolica + media_pressao_mulheres_diastolica) / 2)
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
media_passos_acima_70 = df_saude[df_saude['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_70_ou_menos = df_saude[df_saude['Frequência cardíaca'] <= 70]['Passos diários'].mean()

print ('A média dos passos com pessoas acima dos 70 e com frequência cardiáca maior que 70 é de: ', media_passos_acima_70)
print ('A média dos passos com pessoas acima dos 70 e com frequência cardiáca menor igual a que 70 é de: ', media_passos_70_ou_menos)
```
