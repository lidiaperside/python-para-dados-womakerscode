```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('collected research people.csv')
df.head()
```

```python
#Removendo coluna desnecessária
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.columns
```

```python
#Dicionário de mapeamento para renomear as colunas
novas_cols = {'Endereerço': 'Endereço', 
              'phone': 'Telefone', 
              'anual_income': 'Renda Anual', 
              'AAge': 'Idade', 
              'n_filhos': 'N Filhos',  
              'Food_restrictions': 'Restrição Alimentar', 
              'streaming': 'Streaming', 
              'diariamente_sun blocker': 'Usa protetor solar diariamente', 
              'email': 'Email'}
df.rename(columns=novas_cols, inplace=True)
df.head()
```

```python
df.Nome
```

Utilizando expressão regular

- \W equivale a [^a-zA-Zà-úÀ-Ú0-9_]

```python
#Removendo caracteres especiais da coluna Nome
df['Nome'].str.replace(r'\W', '', regex=True) # == df['Nome'].str.replace(r'[^a-zA-Zà-úÀ-Ú0-9_]', '', regex=True)
```

```python
df['Nome'].str.replace(r'[^a-zA-Zà-úÀ-Ú0-9_]', '', regex=True)
```

```python
#Removendo o _ e adicionando espaço em branco
df['Nome'] = df['Nome'].str.replace(r'[^a-zA-Zà-úÀ-Ú0-9 ]', '', regex=True)
df.head()
```

```python
#Removendo caracteres especiais da coluna Telefone, mas se atentando ao que dee ser mantido
df.Telefone
```

```python
#Removendo tudo que seja no intervalo 0-9, ' ' - + e (), adicionando * nos caracteres especiais como forma de marcação
df.Telefone = df['Telefone'].str.replace(r'[^0-9() +-]', '*', regex=True)
df.Telefone.head()
```

```python
#Substituindo os * por Nan
df.loc[df.Telefone.str.contains('[*]', na=False), 'Telefone'] = np.nan
df.Telefone.head()
```

```python
#Substituindo NaN por Telefone não informado
df.Telefone = df.Telefone.fillna('Telefone não informado')
df.Telefone.head()
```

```python
df['Estado Civil']
```

```python
#Criando um novo dicionário de substituição
df['Estado Civil'].replace({'S': 'Solteiro',
                            'C': 'Casado',
                            'V': 'Viúvo',
                            'D': 'Divorciado'}, inplace = True)
df['Estado Civil'].head()
```

```python
#Substituindo NaN por Estado Civil não informado
df['Estado Civil'].fillna('Estado Civil não informado', inplace=True)
df['Estado Civil'].head()
```

```python
df.head()
```
